```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
eng_stop = set(stopwords.words('english'))
wordlemma = WordNetLemmatizer()
sentences = sent_tokenize(my_data)
for sentence in sentences:
    # 1. Tokenize the sentence into a list of words
    words = word_tokenize(sentence)
    
    # 2. Filter out stopwords and clean each word in a list comprehension
    filtered_words = [word.lower().strip() for word in words if word.lower() not in eng_stop]
    
    # 3. Pass the *list* of words to nltk.pos_tag()
    mywordtuple = nltk.pos_tag(filtered_words)

    # 4. Iterate over the resulting list of tuples and print each one
    for word, tag in mywordtuple:
        print(f"Word: '{word}' | POS Tag: {tag}")  
    
```
